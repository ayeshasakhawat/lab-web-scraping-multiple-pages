```python

from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
# Loading the contents
url = 'https://en.wikipedia.org/wiki/Python'
response = requests.get(url, headers = {"Accept-Language": "en-US"})
response.status_code
```

```python
soup = BeautifulSoup(response.content, 'html.parser')
```

```python
print(soup.prettify())
```

```python
soup.title
```

```python
soup.title.get_text()
```

```python
selector = soup.select("#mw-content-text > div.mw-parser-output a ")
selector
```

```python
links = []
```

```python
# Getting the href out for all links
for item in range(len(selector)):
    links.append(selector[item].attrs['href'])
```

```python
teamLinks = []
```

```python
for i in range(len(selector)):
    teamLinks.append(selector[i].get("href"))
```

```python
teamLinks
```

```python
links_final = []
for item in teamLinks:
    if item.startswith('/'):
        links_final.append('https://en.wiktionary.org'+item)
    elif item.startswith('https:'):
        links_final.append(item)  
```

```python
links_final
```

```python
# Loading the contents
url = 'https://en.wikipedia.org/wiki/List_of_languages_by_number_of_native_speakers'
```

```python
langauageselector = soup1.select("tbody > tr > td > a.mw-redirect")
langauageselector
```

```python
languages =soup1.find('table',{'class':"wikitable"})
languages
```

```python
soup1.select("tbody > tr > td > a.mw-redirect")[0].get_text()
```

```python
langauge = []

langaugelist = soup1.select("tbody > tr > td > a.mw-redirect")
for i in range(len(langauageselector)):
    langauge.append(langaugelist[i].get_text())
```
